```python
def linear_neural_net(X,W, bias):
    yin= sum(x*w for x, w in zip(X,W))+ bias
    if yin < 0:
        output=0
    elif yin > 1:
        output=1
    else:
        output= yin
    return yin, output
```

```python
X=list(map(float, input("X:").split()))
W=list(map(float, input("W:").split()))
b=float(input("B:"))
```

```python
if len(X) != len(W):
    print("Length mismatch")
else:
    yin,out = linear_neural_net(X,W,b)
print("YIN", yin)
print("OUT", out)
```
