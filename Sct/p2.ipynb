```python
print("ANDNOT function using McCulloch-Pitts\n")  # title/header

# inputs for x1,x2
x1 = [1, 1, 0, 0]           # list of x1 values for the four combinations  
x2 = [1, 0, 1, 0]           # list of x2 values for the four combinations  

# ---- Case 1: both weights excitatory ----
print("Case 1: both weights excitatory")  
w1 = [1]*4                 # weight list for x1 (all +1)  
w2 = [1]*4                 # weight list for x2 (all +1)  

# compute weighted sums (yin) for each combination  
yin = [ x1[i]*w1[i] + x2[i]*w2[i] for i in range(4) ]  
print("x1 x2 yin")  
for i in range(4):  
    print(x1[i], x2[i], yin[i])

# ---- Case 2: w1 excitatory, w2 inhibitory ----
print("\nCase 2: w1 excitatory, w2 inhibitory")  
w1 = [1]*4                # same +1 weights for x1  
w2 = [-1]*4               # weights for x2 are −1 (inhibitory)  

# compute new weighted sums
yin = [ x1[i]*w1[i] + x2[i]*w2[i] for i in range(4) ]  
print("x1 x2 yin")  
for i in range(4):  
    print(x1[i], x2[i], yin[i])

# ---- threshold & output decision ----
theta = 2*1 - 1            # threshold (“theta”) calculation (n*w − p)  
print("\nThreshold (theta) =", theta)  
print("Applying threshold")

# apply threshold: if yin ≥ theta → output 1, else 0  
Yin = [ 1 if val >= theta else 0 for val in yin ]  
print("x1 x2 Yin")  
for i in range(4):  
    print(x1[i], x2[i], Yin[i])
```

```python
print("XOR function using McCulloch-Pitts\n")

# Input combinations for XOR
x1 = [1, 1, 0, 0]
x2 = [1, 0, 1, 0]

# ---------------- Step 1: Compute z1 = x1 (excitatory) + x2 (inhibitory) ----------------
print("Calculating z1 = x1*w1 + x2*w2'")
print("Case: x1 = excitatory (+1), x2 = inhibitory (−1)")

w1 = [1]*4          # weights for x1 (+1)
w2 = [-1]*4         # weights for x2 (−1)

z1 = [x1[i]*w1[i] + x2[i]*w2[i] for i in range(4)]

print("x1 x2 z1")
for i in range(4):
    print(x1[i], x2[i], z1[i])

# ---------------- Step 2: Compute z2 = x1' + x2 (invert roles) ----------------
print("\nCalculating z2 = x1' + x2")
print("Case: x1 = inhibitory (−1), x2 = excitatory (+1)")

w1 = [-1]*4         # x1 now inhibitory
w2 = [1]*4          # x2 excitatory

z2 = [x1[i]*w1[i] + x2[i]*w2[i] for i in range(4)]

print("x1 x2 z2")
for i in range(4):
    print(x1[i], x2[i], z2[i])

# ---------------- Step 3: Apply Threshold = 1 to z1 and z2 ----------------
print("\nApplying Threshold = 1 for z1 and z2")

for i in range(4):
    z1[i] = 1 if z1[i] >= 1 else 0
    z2[i] = 1 if z2[i] >= 1 else 0

print("z1 z2")
for i in range(4):
    print(z1[i], z2[i])

# ---------------- Step 4: Output neuron (acts as OR gate) ----------------
print("\nFinal Output y = z1 + z2")

y = []
v1 = 1
v2 = 1

for i in range(4):
    y.append(z1[i]*v1 + z2[i]*v2)

print("x1 x2 y")
for i in range(4):
    print(x1[i], x2[i], y[i])
```
