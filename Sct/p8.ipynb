```python
# Membership Operators
print("=== Membership Operators ===")
fruits = ["apple", "banana", "cherry"]
# Check if "banana" is in the list
if "banana" in fruits:
    print("✅ Banana is in the list")
# Check if "mango" is not in the list
if "mango" not in fruits:
    print("✅ Mango is not in the list")

# Identity Operators
print("\n=== Identity Operators ===")
x = [1, 2, 3]
y = [1, 2, 3]
z = x
# Check if x and z refer to the same object
if x is z:
    print("✅ x and z are the same object (identity)")
# Check if x and y do not refer to the same object
if x is not y:
    print("✅ x and y are not the same object (even if contents are same)")
```

```python
# === Membership Operators ==
print("=== Membership Operators ===")
colors = ["red", "blue", "green"]
# Check if "blue" is in the list
print("blue" in colors) # True
# Check if "yellow" is not in the list
print("yellow" not in colors) # True
# === Identity Operators ===
print("\n=== Identity Operators ===")
a = [10, 20]
b = [10, 20]
c = a
# 'is' checks if two variables point to the same object
print(a is c) # True (same object)
print(a is b) # False (same content, different object)
# 'is not' checks if two variables do not point to the same object
print(a is not b) # True
print(a is not c) # False
```
